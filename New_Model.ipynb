```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
vocab_size = 123 
```

```python
path = '/content/drive/MyDrive/LLM_Training/model-final.pth'
model = GPTLanguageModel(vocab_size)
model.load_state_dict(torch.load(path, map_location=device))
model.to(device)
model.eval()

print("Model loaded successfully!!!")
```

```python
# 1. Load the real vocabulary from your training data
with open('/content/drive/MyDrive/full_book.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    chars = sorted(list(set(text)))

vocab_size = len(chars)
string_to_int = { ch:i for i,ch in enumerate(chars) }
int_to_string = { i:ch for i,ch in enumerate(chars) }

# Updated encode to skip characters not found in the vocabulary to prevent KeyError
encode = lambda s: [string_to_int[c] for c in s if c in string_to_int]
decode = lambda l: ''.join([int_to_string[i] for i in l])

print(f"Vocabulary loaded. Size: {vocab_size}")
print(f"Test: {decode(encode('hello world'))}")
```

```python
# 1. Your prompt
prompt = "Neural Networks and Deep Learning?"

# 2. Convert text to numbers (tensor)
# This will now use the updated 'encode' and won't crash on unknown characters
context = torch.tensor(encode(prompt), dtype=torch.long, device=device).unsqueeze(0)

# 3. Generate new tokens
generated_indices = model.generate(context, max_new_tokens=200, temperature=0.8, top_k=40)

# 4. Convert numbers back to text using our custom decode function
output_text = decode(generated_indices[0].tolist())

print("\n--- Generated Result ---\n")
print(output_text)
```
